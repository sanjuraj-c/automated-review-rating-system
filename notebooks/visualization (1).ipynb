```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load_cleaned data-1
df=pd.read_csv('cleaned_data_1.0.csv')
```

```python
#view
df.head()
```

```python
#histogram of rating
plt.hist(df['Rating'],bins=30)
plt.title('Histogram-Ratings')
plt.show()
```

```python
#load_cleaned data-2
df1=pd.read_csv('cleaned_data_2_0.csv')
```

```python
df1.head()
```

```python
#histogram
sns.histplot(df1['Rating'])
plt.title("Histogram of rating")
plt.show()
```

```python
#combining both cleaned data of both phases
final_combined_df=pd.concat([df,df1],ignore_index=True)
```

```python
#view
final_combined_df.head()
```

```python
#inspect
final_combined_df.shape
```

```python
#check null
final_combined_df.isnull().sum()
```

```python
#check duplicates
final_combined_df.duplicated().sum()
```

```python
#remove duplicates
final_combined_df.drop_duplicates(inplace=True)
```

```python
#histogram
#histogram of rating
plt.hist(final_combined_df['Rating'],bins=30)
plt.title('Histogram-Ratings')
plt.show()
```

```python
#save
final_combined_df.to_csv('final_cleaned.csv')
```

```python
df2=pd.read_csv('final_cleaned.csv')
```

```python
#Count plot (for categorical data)
sns.countplot(x='Rating', data=df2)
plt.title("Rating Class Count")
plt.show()
```

```python
#pie chart
rating_counts = df2['Rating'].value_counts().sort_index()
plt.pie(rating_counts, labels=rating_counts.index, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```
